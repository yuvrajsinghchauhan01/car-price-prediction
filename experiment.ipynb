```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cars_24_combined.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Year']=df['Year'].astype('int')
```

```python
df.head()
```

```python
# get current year
from datetime import date

current_year = date.today().year
current_year
```

```python
df['car_age'] = current_year - df['Year']
df.head()
```

```python
df.drop('Year', axis=1, inplace=True)
```

```python
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
X.head()
```

```python
X.columns
```

```python
X['Drive'].value_counts()
```

```python
X['Type'].value_counts().index
```

```python
# split columns in numeric and categorical
df_num = X.select_dtypes(include=['int64', 'float64'])
df_cat = X.select_dtypes(include=['object'])
```

```python
df_num.head()
```

```python
df_cat.head()
```

```python
# scaling the numeric columns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_num_scaled = scaler.fit_transform(df_num)
df_num_scaled = pd.DataFrame(df_num_scaled, columns=df_num.columns)
df_num_scaled.head()
```

```python
# one-hot encoding the categorical columns
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
df_cat_encoded = encoder.fit_transform(df_cat)
df_cat_encoded = pd.DataFrame(df_cat_encoded.toarray(), columns=encoder.get_feature_names_out())
df_cat_encoded.head()
```

```python
# concatenate the numeric and categorical columns
X_encoded = pd.concat([df_num_scaled, df_cat_encoded], axis=1)
X_encoded.head()
```

```python
# save X_encoded as a CSV file
X.to_csv('X.csv', index=False)
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# predict on test set
y_pred = model.predict(X_test)
```

```python
# evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE:', mse)
print('R2 score:', r2)
```

```python
# save the model
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
